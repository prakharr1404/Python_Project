```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import ast 

df = pd.read_csv(r"C:\Users\prakh\OneDrive\Desktop\Python\data_jobs.csv")

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])

df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
df[['job_title_short', 'job_skills']].head(5)
```

```python
help(df.explode)
```

```python
df_exploded = df.explode('job_skills')
df_exploded
```

```python
#sample code, not part of actual file
df_exploded['job_skills'].value_counts().head(10).plot(kind='bar')
```

```python
skills_count = df_exploded.groupby(by=['job_title_short', 'job_skills']).size()
type(skills_count)
```

```python
df_skills_count = skills_count.reset_index(name='skill_count')
df_skills_count
```

```python
df_skills_count = df_skills_count.sort_values(by='skill_count', ascending=False)
df_skills_count
```

```python
job_title = 'Data Analyst'
top_skills = 10

df_skills_final = df_skills_count[df_skills_count['job_title_short'] == job_title].head(top_skills)
df_skills_final
                                                                                        
```

```python
df_skills_final.plot(kind='bar', x='job_skills', y='skill_count')
plt.legend().set_visible(False)
plt.title(f'Top {top_skills} Skills For {job_title}')
```
